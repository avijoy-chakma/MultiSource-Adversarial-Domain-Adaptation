## Convert All the File Timestamp into EST

```python
def ReadShimmer(filename):

    RAW = pd.read_csv(filename,skiprows=3, sep=",", header=None)
    RAW.columns = ["Timestamp", "Accel_LN_X_CAL", "Accel_LN_Y_CAL", "Accel_LN_Z_CAL", "Accel_WR_X_CAL" ,"Accel_WR_Y_CAL","Accel_WR_Z_CAL",
             "Gyro_X_CAL","Gyro_Y_CAL","Gyro_Z_CAL","Mag_X_CAL","Mag_Y_CAL","Mag_Z_CAL"]

    Timestamp=(RAW['Timestamp'].astype(float)).values
    Time_stamp = np.vectorize(datetime.utcfromtimestamp)(np.array(Timestamp/1000.0)).astype('datetime64[us]')
    RAW['Timestamp'] = Time_stamp
    
    Timestamps=pd.to_datetime(RAW['Timestamp'],unit='s', utc=True).dt.tz_convert('US/Eastern')
    RAW['Timestamp'] = Timestamps

    print('Generated Dataframe')
    return RAW


def plot_activity(df, position='Default'):
    data = df[['AccX', 'AccY','AccZ']][:df.shape[0]]

#     position += '  Index Range: '+str(df.index[0])+ '  to  '+ str(df.index[-1])
    axis = data.plot(subplots=True, figsize=(16, 12), title=position)
    for ax in axis:
        ax.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
#         ax.figure.savefig(visualization_url+position+'.png')
```

## Split into Train and Valid Files

```python
def split_files(dataframe, position, activity_index):
    
    train_dataframe = pd.DataFrame(columns=['Timestamp', 'Accel_LN_X_CAL', 'Accel_LN_Y_CAL', 'Accel_LN_Z_CAL', 'gt', 'position'])
    valid_dataframe = pd.DataFrame(columns=['Timestamp', 'Accel_LN_X_CAL', 'Accel_LN_Y_CAL', 'Accel_LN_Z_CAL', 'gt', 'position'])
    
    upto = int(dataframe.shape[0]*split_percentage) 
    train = dataframe[0:upto].copy()
    valid = dataframe[upto+1 : dataframe.shape[0]].copy()

    train_dataframe = train_dataframe.append(train, ignore_index=True, sort=False)
    valid_dataframe = valid_dataframe.append(valid, ignore_index=True, sort=False)
        
    train_dataframe.to_csv (save_path+user[0]+"_"+position+'_'+activity[activity_index]+'_train'+".csv", index = None, header=True)
    valid_dataframe.to_csv (save_path+user[0]+"_"+position+'_'+activity[activity_index]+'_valid'+".csv", index = None, header=True)
```

```python
def standardization_and_normalization(dataframe):
    timestamp_gt = np.array(dataframe['Timestamp']) 
    activity_gt = np.array(dataframe['gt'])
    position_gt = np.array(dataframe['position'])

    dataframe.drop(['Timestamp','gt','position'], axis=1, inplace=True)
    column_name = dataframe.columns

    scaler = preprocessing.StandardScaler()
    df_standardized = scaler.fit_transform(dataframe)
    df_standardized = pd.DataFrame(df_standardized)

    min_max_scaler = preprocessing.MinMaxScaler()
    np_scaled = min_max_scaler.fit_transform(df_standardized)
    dataframe = pd.DataFrame(np_scaled, columns=column_name)

    dataframe["Activity"] = activity_gt
    dataframe["Body_Position"] = position_gt
    
    return dataframe
```

```python
for index in range(0,5):
    split_files(walking[index], position[index], 0)
    split_files(sitting[index], position[index], 1)
    split_files(running[index], position[index], 2)
    split_files(biking[index], position[index], 3)
```

```python
train = {}
valid = {}
```

```python
for user_index in range(0,1):
    for position_index in range(0,5):
        
#         train_dataframe = pd.DataFrame(columns=['Timestamp', 'Accel_LN_X_CAL', 'Accel_LN_Y_CAL', 'Accel_LN_Z_CAL', 'gt', 'position'])
#         valid_dataframe = pd.DataFrame(columns=['Timestamp', 'Accel_LN_X_CAL', 'Accel_LN_Y_CAL', 'Accel_LN_Z_CAL', 'gt', 'position'])
        
        train_dataframe = pd.DataFrame()
        valid_dataframe = pd.DataFrame()
    
        for activity_index in range(0,4):
            train_file = user[user_index]+ "_" + position[position_index] + "_" + activity[activity_index] + "_train.csv"
            valid_file = user[user_index]+ "_" + position[position_index] + "_" + activity[activity_index] + "_valid.csv"
            
            train_frame = pd.read_csv(save_path + train_file, skiprows = 1, header=None)
            valid_frame = pd.read_csv(save_path + valid_file, skiprows = 1, header=None)
            
            train_dataframe = train_dataframe.append(train_frame, ignore_index=True, sort=False)
            valid_dataframe = valid_dataframe.append(valid_frame, ignore_index=True, sort=False)
        
        train_dataframe.rename(columns = {0:'Timestamp', 1:'Accel_LN_X_CAL', 2:'Accel_LN_Y_CAL', 3:'Accel_LN_Z_CAL', 4:'gt', 5:'position'}, inplace = True) 
        valid_dataframe.rename(columns = {0:'Timestamp', 1:'Accel_LN_X_CAL', 2:'Accel_LN_Y_CAL', 3:'Accel_LN_Z_CAL', 4:'gt', 5:'position'}, inplace = True) 
        
#         print(train_dataframe.head())
        train_dataframe = standardization_and_normalization(train_dataframe)
        valid_dataframe = standardization_and_normalization(valid_dataframe)
        
        train[position_index] = train_dataframe
        valid[position_index] = valid_dataframe
        
        train[position_index].to_csv (processed_file_path+user[user_index]+"_"+position[position_index]+'_train'+".csv", index = None, header=True)
        valid[position_index].to_csv (processed_file_path+user[user_index]+"_"+position[position_index]+'_valid'+".csv", index = None, header=True)
        
        if position_index == 0:
            plot_activity(train[position_index], "Chest")
            plot_activity(valid[position_index], "Chest")
```

## Preparing the Windowed Dataset

```python
source1 = 1
source2 = 2
target = 3
selected_user = 1
item = ["train","valid"]
```

```python
win_size=256
overlap = 0.5
step_size=int(win_size * overlap)

AXIS = 3
FROM = 0
TO = FROM+3
START = 3
END = 4
```

```python
chest_train = []
right_hand_train = []
right_leg_train = []
left_hand_train = []
left_leg_train = []

chest_train_gt = []
right_hand_train_gt = []
right_leg_train_gt = []
left_hand_train_gt = []
left_leg_train_gt = []
```

```python
chest_valid = []
right_hand_valid = []
right_leg_valid = []
left_hand_valid = []
left_leg_valid = []

chest_valid_gt = []
right_hand_valid_gt = []
right_leg_valid_gt = []
left_hand_valid_gt = []
left_leg_valid_gt = []
```

```python
for position_index in tqdm.tqdm(range(0,5)):
    for split_index in range(0,2):
        file_name = user[selected_user-1] + "_" + position[position_index]+'_'+item[split_index]
        
        print(file_name)
        df = pd.read_csv(processed_file_path+file_name+'.csv', sep=",")   
        len_df = df.shape[0]
        narray = df.to_numpy()
        
        for i in range(0, len_df, step_size):
            window = narray[i:i+win_size, FROM:TO]
            
            if window.shape[0] != win_size:
                continue
            else:
                reshaped_window = window.reshape(1,win_size,1,AXIS)
                gt = np.bincount(narray[i:i+win_size,START:END].astype(int).ravel()).argmax()
                
                if position_index == 0:
                    if split_index == 0:
                        chest_train.append(reshaped_window)
                        chest_train_gt.append(gt)
                    elif split_index == 1:
                        chest_valid.append(reshaped_window)
                        chest_valid_gt.append(gt)
                elif position_index == 1:
                    if split_index == 0:
                        right_hand_train.append(reshaped_window)
                        right_hand_train_gt.append(gt)
                    elif split_index == 1:
                        right_hand_valid.append(reshaped_window)
                        right_hand_valid_gt.append(gt)
                elif position_index == 2:
                    if split_index == 0:
                        right_leg_train.append(reshaped_window)
                        right_leg_train_gt.append(gt)
                    elif split_index == 1:
                        right_leg_valid.append(reshaped_window)
                        right_leg_valid_gt.append(gt)
                elif position_index == 3:
                    if split_index == 0:
                        left_hand_train.append(reshaped_window)
                        left_hand_train_gt.append(gt)
                    elif split_index == 1:
                        left_hand_valid.append(reshaped_window)
                        left_hand_valid_gt.append(gt)
                elif position_index == 4:
                    if split_index == 0:
                        left_leg_train.append(reshaped_window)
                        left_leg_train_gt.append(gt)
                    elif split_index == 1:
                        left_leg_valid.append(reshaped_window)
                        left_leg_valid_gt.append(gt)
```
