```python
total_frame = {}
```

```python
for user_index in range(1,5):
    for run_index in range(1,6):
        file_name = "S"+str(user_index)+"-ADL"+str(run_index)+".dat"
        dataframe = pd.read_csv(url+file_name,sep=" ", header=None) 
        
        BACK_IMU_frame = dataframe[[37,38,39,40,41,42,43,44,45]].copy()
        BACK_IMU_frame.rename(columns={37: 'BACK_Acc_x', 38:'BACK_Acc_y', 39: 'BACK_Acc_z', 40: 'BACK_Gyr_x', 41:'BACK_Gyr_y', 42: 'BACK_Gyr_z', 43: 'BACK_Mag_x', 44:'BACK_Mag_y', 45: 'BACK_Mag_z'}, inplace=True)
        
        RUA_IMU_frame = dataframe[[50,51,52,53,54,55,56,57,58]].copy()
        RUA_IMU_frame.rename(columns={50: 'RUA_Acc_x', 51:'RUA_Acc_y', 52: 'RUA_Acc_z', 53: 'RUA_Gyr_x', 54:'RUA_Gyr_y', 55: 'RUA_Gyr_z', 56: 'RUA_Mag_x', 57:'RUA_Mag_y', 58: 'RUA_Mag_z'}, inplace=True)
        
        RLA_IMU_frame = dataframe[[63,64,65,66,67,68,69,70,71]].copy()
        RLA_IMU_frame.rename(columns={63: 'RLA_Acc_x', 64:'RLA_Acc_y', 65: 'RLA_Acc_z', 66: 'RLA_Gyr_x', 67:'RLA_Gyr_y', 68: 'RLA_Gyr_z', 69: 'RLA_Mag_x', 70:'RLA_Mag_y', 71: 'RLA_Mag_z'}, inplace=True)
        
        LUA_IMU_frame = dataframe[[76,77,78,79,80,81,82,83,84]].copy()
        LUA_IMU_frame.rename(columns={76: 'LUA_Acc_x', 77:'LUA_Acc_y', 78: 'LUA_Acc_z', 79: 'LUA_Gyr_x', 80:'LUA_Gyr_y', 81: 'LUA_Gyr_z', 82: 'LUA_Mag_x', 83:'LUA_Mag_y', 84: 'LUA_Mag_z'}, inplace=True)
        
        LLA_IMU_frame = dataframe[[89,90,91,92,93,94,95,96,97]].copy()
        LLA_IMU_frame.rename(columns={89: 'LLA_Acc_x', 90:'LLA_Acc_y', 91: 'LLA_Acc_z', 92: 'LLA_Gyr_x', 93:'LLA_Gyr_y', 94: 'LLA_Gyr_z', 95: 'LLA_Mag_x', 96:'LLA_Mag_y', 97: 'LLA_Mag_z'}, inplace=True)
        
        gt = dataframe[[243]].copy()
        # Merge into a frame
        merged_frame = pd.concat([BACK_IMU_frame, RUA_IMU_frame, RLA_IMU_frame, LUA_IMU_frame, LLA_IMU_frame], axis = 1)
        merged_frame.loc[:, 'Person'] = user_index
        merged_frame["gt"] = gt
        total_frame[user_index] = merged_frame
        total_frame[user_index].reset_index(drop=True)
        
```

## Prepare Final Dataset

```python
def initialization():
    back_dataset = []
    RUA_dataset = []
    RLA_dataset = []
    LUA_dataset = []
    LLA_dataset = []
    back_gt = []
    RUA_gt = []
    RLA_gt = []
    LUA_gt = []
    LLA_gt = []
```

```python
position = ["back","RUA","RLA","LUA","LLA"]
```

```python
win_size=128
step_size=64
```

```python
initialization()
for position_index in tqdm(range(1,6)): #Back, RUA, RLA, LUA, LLA
    for person_index in range(1,5): # 4 Persons
        file_name = "user"+str(person_index) + "_"+position[position_index-1]
        
        df = pd.read_csv(save_path+file_name+'.csv', sep=",")   
        len_df = df.shape[0]
        narray = df.to_numpy()

        for i in range(0, len_df, step_size):
            window = narray[i:i+win_size,0:9]
            
            if reshaped_window.shape[0] != win_size:
                continue
            else:
                reshaped_window = window.reshape(1,win_size,1,9)
                gt = np.bincount(narray[i:i+win_size,9:10].astype(int).ravel()).argmax()
                print(gt)
                
                if position_index == 1:
                    back_dataset.append(reshaped_window)
                    back_gt.append(gt)
                elif position_index == 2:
                    RUA_dataset.append(reshaped_window)
                    RUA_gt.append(gt)
                elif position_index == 3:
                    RLA_dataset.append(reshaped_window)
                    RLA_gt.append(gt)
                elif position_index == 4:
                    LUA_dataset.append(reshaped_window)
                    LUA_gt.append(gt)
                elif position_index == 5:
                    LLA_dataset.append(reshaped_window)
                    LLA_gt.append(gt)
```
